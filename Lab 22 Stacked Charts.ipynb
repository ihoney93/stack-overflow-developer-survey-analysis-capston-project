### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
# Group by Age and calculate median for both satisfaction columns
medians = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Draw stacked bar chart
medians.plot(kind='bar', stacked=True)
plt.title("Job Satisfaction by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median Satisfaction Score")
plt.xticks(rotation=90)
plt.legend(['JobSatPoints_6', 'JobSatPoints_7'])
plt.tight_layout()
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
medians = df.groupby('Employment')[['JobSatPoints_6', 'JobSatPoints_7']].median()

medians.plot(kind='bar', stacked=True)
plt.title("Job Satisfaction by Employment Status")
plt.xlabel("Employment Type")
plt.ylabel("Median Satisfaction Score")
plt.xticks(rotation=90)
plt.legend(['JobSatPoints_6', 'JobSatPoints_7'])
plt.tight_layout()
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
# Turn Age into numbers
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': 25
}
df['Age_numeric'] = df['Age'].map(age_map)

# Filter for ages 30-35
df_filtered = df[df['Age'].isin(['25-34 years old', '35-44 years old'])]

# Group by Age and calculate median
medians = df_filtered.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

medians.plot(kind='bar', stacked=True)
plt.title("Compensation and Job Satisfaction (Ages 30-35)")
plt.xlabel("Age Group")
plt.ylabel("Median Value")
plt.xticks(rotation=0)
plt.legend(['Compensation', 'Job Satisfaction'])
plt.tight_layout()
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
medians = df.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

medians.plot(kind='bar', stacked=True)
plt.title("Compensation and Job Satisfaction by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median Value")
plt.xticks(rotation=90)
plt.legend(['Compensation', 'Job Satisfaction'])
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
# Split databases and explode
df_db = df[['DatabaseWantToWorkWith', 'Age']].dropna()
df_db = df_db.assign(
    DatabaseWantToWorkWith=df_db['DatabaseWantToWorkWith'].str.split(';')
).explode('DatabaseWantToWorkWith')

# Keep top 5 databases only
top5_db = df_db['DatabaseWantToWorkWith'].value_counts().head(5).index
df_db = df_db[df_db['DatabaseWantToWorkWith'].isin(top5_db)]

# Create a cross-tab (counts each database + age combo)
ct = pd.crosstab(df_db['Age'], df_db['DatabaseWantToWorkWith'])

# Draw stacked bar chart
ct.plot(kind='bar', stacked=True)
plt.title("Preferred Databases by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.legend(title="Database", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
# Drop empty rows
df_emp = df[['Employment', 'JobSatPoints_6']].dropna()

# Create cross-tab
ct = pd.crosstab(df_emp['JobSatPoints_6'], df_emp['Employment'])

# Draw stacked bar chart
ct.plot(kind='bar', stacked=True)
plt.title("Employment Type by Job Satisfaction")
plt.xlabel("Job Satisfaction Score")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Employment", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
# Split languages and explode
df_lang = df[['LanguageWantToWorkWith', 'Age']].dropna()
df_lang = df_lang.assign(
    LanguageWantToWorkWith=df_lang['LanguageWantToWorkWith'].str.split(';')
).explode('LanguageWantToWorkWith')

# Keep top 5 languages
top5_lang = df_lang['LanguageWantToWorkWith'].value_counts().head(5).index
df_lang = df_lang[df_lang['LanguageWantToWorkWith'].isin(top5_lang)]

# Cross-tab
ct = pd.crosstab(df_lang['Age'], df_lang['LanguageWantToWorkWith'])

# Draw stacked bar chart
ct.plot(kind='bar', stacked=True)
plt.title("Preferred Languages by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.legend(title="Language", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
# Split platforms and explode
df_plat = df[['PlatformHaveWorkedWith', 'Employment']].dropna()
df_plat = df_plat.assign(
    PlatformHaveWorkedWith=df_plat['PlatformHaveWorkedWith'].str.split(';')
).explode('PlatformHaveWorkedWith')

# Keep top 5 platforms
top5_plat = df_plat['PlatformHaveWorkedWith'].value_counts().head(5).index
df_plat = df_plat[df_plat['PlatformHaveWorkedWith'].isin(top5_plat)]

# Cross-tab
ct = pd.crosstab(df_plat['Employment'], df_plat['PlatformHaveWorkedWith'])

# Draw stacked bar chart
ct.plot(kind='bar', stacked=True)
plt.title("Platform Adoption by Employment Type")
plt.xlabel("Employment Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.legend(title="Platform", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```
