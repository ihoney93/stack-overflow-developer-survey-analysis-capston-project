### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
# Turn Age into numbers
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': 25
}
df['Age_numeric'] = df['Age'].map(age_map)

# Turn SOPartFreq into numbers
freq_map = {
    'Multiple times per day': 5,
    'Daily or almost daily': 4,
    'A few times per week': 3,
    'A few times per month or weekly': 2,
    'Less than once per month or monthly': 1
}
df['SOPartFreq_numeric'] = df['SOPartFreq'].map(freq_map)

# Drop empty rows
df_clean = df.dropna(subset=['Age_numeric', 'SOPartFreq_numeric', 'JobSat'])

# Draw bubble plot
plt.scatter(x=df_clean['Age_numeric'], 
            y=df_clean['SOPartFreq_numeric'], 
            s=df_clean['JobSat'] * 5, 
            alpha=0.3)
plt.title("Age vs Participation Frequency (Bubble Size = Job Satisfaction)")
plt.xlabel("Age")
plt.ylabel("Participation Frequency")
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
# Remove compensation outliers
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df_comp = df[(df['ConvertedCompYearly'] >= lower) & (df['ConvertedCompYearly'] <= upper)]

# Drop empty rows
df_comp = df_comp.dropna(subset=['ConvertedCompYearly', 'JobSat', 'Age_numeric'])

# Draw bubble plot
plt.scatter(x=df_comp['ConvertedCompYearly'], 
            y=df_comp['JobSat'], 
            s=df_comp['Age_numeric'] * 3, 
            alpha=0.3)
plt.title("Compensation vs Job Satisfaction (Bubble Size = Age)")
plt.xlabel("Yearly Compensation")
plt.ylabel("Job Satisfaction")
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
# Split languages and explode
df_lang = df[['LanguageHaveWorkedWith', 'Age']].dropna()
df_lang = df_lang.assign(
    LanguageHaveWorkedWith=df_lang['LanguageHaveWorkedWith'].str.split(';')
).explode('LanguageHaveWorkedWith')

# Map age to numbers
df_lang['Age_numeric'] = df_lang['Age'].map(age_map)

# Get top 10 languages
top10 = df_lang['LanguageHaveWorkedWith'].value_counts().head(10).index
df_lang = df_lang[df_lang['LanguageHaveWorkedWith'].isin(top10)]

# Count frequency for each language + age combo
counts = df_lang.groupby(['LanguageHaveWorkedWith', 'Age_numeric']).size().reset_index(name='count')

# Draw bubble plot
plt.scatter(x=counts['Age_numeric'], 
            y=counts['LanguageHaveWorkedWith'], 
            s=counts['count'] * 2, 
            alpha=0.5)
plt.title("Programming Languages by Age Group")
plt.xlabel("Age")
plt.ylabel("Language")
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
# Split databases and explode
df_db = df[['DatabaseWantToWorkWith', 'JobSat']].dropna()
df_db = df_db.assign(
    DatabaseWantToWorkWith=df_db['DatabaseWantToWorkWith'].str.split(';')
).explode('DatabaseWantToWorkWith')

# Get top 10 databases
top10_db = df_db['DatabaseWantToWorkWith'].value_counts().head(10).index
df_db = df_db[df_db['DatabaseWantToWorkWith'].isin(top10_db)]

# Calculate average satisfaction and count for each database
db_stats = df_db.groupby('DatabaseWantToWorkWith').agg(
    avg_sat=('JobSat', 'mean'),
    count=('JobSat', 'size')
).reset_index()

# Draw bubble plot
plt.scatter(x=db_stats['avg_sat'], 
            y=db_stats['DatabaseWantToWorkWith'], 
            s=db_stats['count'] / 5, 
            alpha=0.6)
plt.title("Preferred Databases vs Job Satisfaction")
plt.xlabel("Average Job Satisfaction")
plt.ylabel("Database")
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
# Split DevType and explode
df_dev = df[['DevType', 'ConvertedCompYearly', 'JobSat']].dropna()
df_dev = df_dev.assign(DevType=df_dev['DevType'].str.split(';')).explode('DevType')

# Get top 10 dev types
top10_dev = df_dev['DevType'].value_counts().head(10).index
df_dev = df_dev[df_dev['DevType'].isin(top10_dev)]

# Remove compensation outliers
Q1 = df_dev['ConvertedCompYearly'].quantile(0.25)
Q3 = df_dev['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df_dev = df_dev[(df_dev['ConvertedCompYearly'] >= lower) & (df_dev['ConvertedCompYearly'] <= upper)]

# Calculate average comp and satisfaction for each role
dev_stats = df_dev.groupby('DevType').agg(
    avg_comp=('ConvertedCompYearly', 'mean'),
    avg_sat=('JobSat', 'mean'),
    count=('JobSat', 'size')
).reset_index()

# Draw bubble plot
plt.scatter(x=dev_stats['avg_comp'], 
            y=dev_stats['DevType'], 
            s=dev_stats['avg_sat'] * 50, 
            alpha=0.6)
plt.title("Compensation by Developer Role (Bubble Size = Satisfaction)")
plt.xlabel("Average Yearly Compensation")
plt.ylabel("Developer Role")
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
# Split tools and explode
df_tools = df[['NEWCollabToolsHaveWorkedWith', 'Age']].dropna()
df_tools = df_tools.assign(
    NEWCollabToolsHaveWorkedWith=df_tools['NEWCollabToolsHaveWorkedWith'].str.split(';')
).explode('NEWCollabToolsHaveWorkedWith')

# Map age
df_tools['Age_numeric'] = df_tools['Age'].map(age_map)

# Get top 10 tools
top10_tools = df_tools['NEWCollabToolsHaveWorkedWith'].value_counts().head(10).index
df_tools = df_tools[df_tools['NEWCollabToolsHaveWorkedWith'].isin(top10_tools)]

# Count each tool + age combo
tool_counts = df_tools.groupby(['NEWCollabToolsHaveWorkedWith', 'Age_numeric']).size().reset_index(name='count')

# Draw bubble plot
plt.scatter(x=tool_counts['Age_numeric'], 
            y=tool_counts['NEWCollabToolsHaveWorkedWith'], 
            s=tool_counts['count'] * 2, 
            alpha=0.5)
plt.title("Collaboration Tools by Age Group")
plt.xlabel("Age")
plt.ylabel("Tool")
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
# Split frameworks and explode
df_web = df[['WebframeWantToWorkWith', 'JobSat']].dropna()
df_web = df_web.assign(
    WebframeWantToWorkWith=df_web['WebframeWantToWorkWith'].str.split(';')
).explode('WebframeWantToWorkWith')

# Get top 10 frameworks
top10_web = df_web['WebframeWantToWorkWith'].value_counts().head(10).index
df_web = df_web[df_web['WebframeWantToWorkWith'].isin(top10_web)]

# Calculate stats
web_stats = df_web.groupby('WebframeWantToWorkWith').agg(
    avg_sat=('JobSat', 'mean'),
    count=('JobSat', 'size')
).reset_index()

# Draw bubble plot
plt.scatter(x=web_stats['avg_sat'], 
            y=web_stats['WebframeWantToWorkWith'], 
            s=web_stats['count'] / 5, 
            alpha=0.6)
plt.title("Web Frameworks vs Job Satisfaction")
plt.xlabel("Average Job Satisfaction")
plt.ylabel("Web Framework")
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
# Split languages and explode
df_admired = df[['LanguageWantToWorkWith', 'Country']].dropna()
df_admired = df_admired.assign(
    LanguageWantToWorkWith=df_admired['LanguageWantToWorkWith'].str.split(';')
).explode('LanguageWantToWorkWith')

# Keep top 5 countries and top 5 languages
top5_countries = df_admired['Country'].value_counts().head(5).index
top5_langs = df_admired['LanguageWantToWorkWith'].value_counts().head(5).index

df_admired = df_admired[
    (df_admired['Country'].isin(top5_countries)) & 
    (df_admired['LanguageWantToWorkWith'].isin(top5_langs))
]

# Count each language + country combo
admired_counts = df_admired.groupby(['LanguageWantToWorkWith', 'Country']).size().reset_index(name='count')

# Draw bubble plot
plt.scatter(x=admired_counts['Country'], 
            y=admired_counts['LanguageWantToWorkWith'], 
            s=admired_counts['count'] * 3, 
            alpha=0.5)
plt.title("Admired Languages Across Countries")
plt.xlabel("Country")
plt.ylabel("Language")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
