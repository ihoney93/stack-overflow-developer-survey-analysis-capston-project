### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
df['Country'].value_counts()
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
df['Country'] = df['Country'].replace({
    'United States of America':'USA',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom'
})
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
df['Employment'].value_counts()
```

```python
pd.get_dummies(df['Employment']).head(10)
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
missing_values = df.isnull().sum().sort_values(ascending=False).head(20)
print(f"The number of missing values: {missing_values}")
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
ConvertedCompYearly_missingvalue = df['ConvertedCompYearly'].sum()
print(f"In the ConvertedCompYearly Column, the number of missing values is: {ConvertedCompYearly_missingvalue}")
```

```python
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode()[0])
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
min_value = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min())
max_value = (df['ConvertedCompYearly'].max()) - (df['ConvertedCompYearly'].min())

df['ConvertedCompYearly_MinMax_Value'] = (min_value) / (max_value)

df['ConvertedCompYearly_MinMax_Value'].head(10)
```

```python
df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax_Value']].head(10)
```

```python
df['ConvertedCompYearly_MinMax_Value'].describe()
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
import numpy as np
```

```python
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'])
```

```python
df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head(10)
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
df['YearsCodePro'].describe()
```

```python
df['YearsCodePro'].value_counts(ascending=True)
```

```python
missing_years_value = df['YearsCodePro'].isnull().sum()
print(f"The number of missing values in YearsCodePro : {missing_years_value}")
```

```python
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
```

```python
df['ExperienceLevel'] = pd.cut(
    df['YearsCodePro'],
    bins=[0,5,10,100], 
    labels=['Junior', 'Mid-level', 'Senior']
)
```

```python
df['ExperienceLevel'].value_counts()
```
